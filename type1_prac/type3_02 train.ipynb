## 02. 독립 표본 검정

스마트폰 배터리의 충전 시간을 줄이기 위한 새로운 충전기를 개발했다. 개발된 충전기의 효과를 검정하기 위해 스마트폰 사용자 집단을 두 그룹으로 나눠 한 그룹에는 새로운 충전기를 제공하고, 다른 그룹에는 기존의 충전기를 제공했다. 그 후 두 그룹의 평균 충전 완료 시간을 비교했다. 두 그룹 모두 충전 시간의 정규 분포를 따르며 분산은 같다고 가정한다. 새로운 충전기가 더 빠르다고 할 수 있는지를 유의수준 5%하에서 검정하시오. 여기서 두 집단은 모두 정규 분포를 따르고 분산은 같다고 가정한다. * 귀무가설($H_0$): $\mu1$ >= $\mu2$
* 대립가설($H_1$): $\mu1$ < $\mu2$
